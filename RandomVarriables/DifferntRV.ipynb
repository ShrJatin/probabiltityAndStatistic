```python
#@title  #Binomial RV
import matplotlib.pyplot as ok
from scipy.stats import uniform
import numpy as np

def Binomial(n, p = 0.4, m = 1000):
  binomial = np.empty(m, dtype=int)

  for i in range(m):
    data_uniform = uniform.rvs(size=n, loc=0, scale=1)
    bernoulli = np.where(data_uniform < p, 1, 0)
    binomial[i] = np.sum(bernoulli)
  rv = binomial
  bin_edges = np.arange(rv.min()-0.5, rv.max()+1.5, 1)
  ok.hist(rv, bins = bin_edges, rwidth = 0.8)
  ok.xticks(np.arange(rv.min(), rv.max()+1, 1))
  ok.xlabel('value')
  ok.ylabel('frequency')
  ok.show()
```

1.1

```python
Binomial(10)
```

1.2

```python
Binomial(1000)
```

1.3

```python
Binomial(200)
```

```python
#@title  Poisson RV
from scipy.stats import poisson
import matplotlib.pyplot as ok
import numpy as np

def pois(alpha):
  rv = poisson.rvs(alpha, size =  1000)
  bin_edges = np.arange(rv.min()-0.5, rv.max()+1.5, 1)

  ok.hist(rv, bins = bin_edges, rwidth = 0.8)
  ok.xticks(np.arange(rv.min(), rv.max()+1, 1))
  ok.xlabel('value')
  ok.ylabel('frequency')
  ok.show()
```

```python
pois(4.5)
```

2.2

```python
pois(500)
```

2.3

```python
pois(30)
```

## Gaussian

```python
#@title Gaussian
import matplotlib.pyplot as ok
import numpy as np
import random

def gauss(mu, sig):
  nums=[]
  for i in range(1000): 
      nums.append(random.gauss(mu, sig)) 

  ok.hist(nums, bins= 200) 
  ok.xlabel('value')
  ok.ylabel('frequency')
  ok.show()
```

3.1

```python
gauss(0, 2)
```

3.2

```python
gauss(10, 5)
```

3.3

```python
gauss(50, 25)
```

```python
#@title Exponential
import matplotlib.pyplot as ok
from numpy import random 
 
size=1000
 
def Exponential(lda):
  data = random.exponential(1.0/lda, size=size)
  ok.hist(data, bins = 100) 
  ok.xlabel('value')
  ok.ylabel('frequency')
  ok.show()
```

```python
Exponential(2)
```

```python
Exponential(25)
```

```python
Exponential(100)
```
