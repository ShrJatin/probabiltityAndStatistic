```python
#@title Write a function that generates m samples from an exponential RV with the rate parameter as input. { display-mode: "form" }
import matplotlib.pyplot as plt
from numpy import random 

# print(data)
Y = []

n = 3
lda = 1/2


for i in range(1000):
  # generate 3 samples of exponential rv
  data = random.exponential(1/lda, size=3)
  # generate a RV by summing up all 3 samples and store in Y
  Y.append(sum(data))

# plot Y
plt.hist(Y, bins = 200)
plt.xlabel('value')
plt.ylabel('frequency')
plt.show()
```

```python
#@title Erlang RV
from scipy.stats import erlang
import matplotlib.pyplot as pl
n = 3
lda = 1/2
y = erlang.rvs(n, loc = 0, scale = 1/lda, size = 1000, random_state = 3)

plt.hist(y, bins = 200)
plt.xlabel('value')
plt.ylabel('frequency')
plt.show()
```

```python
#generate m (say 100) independent samples of Binomial RV with any parameters

from scipy.stats import uniform
import matplotlib.pyplot as plt
import numpy as np

# gnerate sum of 100 binomial variable generator
def binrvs(n, p):
  binomial = np.empty(100, dtype=int)

  for i in range(100):
    data_uniform = uniform.rvs(size=n, loc=0, scale=1)
    bernoulli = np.where(data_uniform < p, 1, 0)
    binomial[i] = np.sum(bernoulli)

  return sum(binomial)

# generate say 1000 samples of Y.
folks = np.empty(1000, dtype=int)
for i in range(1000):
  folks[i]  = binrvs(69, 0.0145)

# plot
bin_edges = np.arange(folks.min()-0.5, folks.max()+1.5, 1)
plt.hist(folks, bins = bin_edges, rwidth = 0.8)
plt.xticks(np.arange(folks.min(), folks.max()+1, 0.5))
plt.xlabel('value')
plt.ylabel('frequency')
plt.show()
```
